```python
data1["MSRP"] = data1['MSRP'].str.strip('$')
data1["Invoice"] = data1["Invoice"].str.strip('$')
data1["MSRP"] = data1["MSRP"].str.replace(',','.')
data1["Invoice"] = data1["Invoice"].str.replace(',','.')
```

```python
data1['MSRP'] = data1['MSRP'] = data1['MSRP'].astype(float)
data1['Invoice'] = data1['Invoice'].astype(float)
```

```python
data1['Make'].fillna("UNKNOWN" , inplace = True)
data1['Model'].fillna("UNKNOWN" , inplace = True)
data1['Type'].fillna("UNKNOWN" , inplace = True)
data1['Origin'].fillna("UNKNOWN" , inplace = True)
data1['DriveTrain'].fillna("UNKNOWN" , inplace = True)
data1['MSRP'].fillna(data1['MSRP'].mean(), inplace = True)
data1['Invoice'].fillna(data1['Invoice'].mean(), inplace = True)
data1['EngineSize'].fillna(data1['EngineSize'].mean(), inplace = True)
data1['Cylinders'].fillna(data1['Cylinders'].mean(), inplace = True)
data1['Horsepower'].fillna(data1['Horsepower'].mean(), inplace = True)
data1['MPG_City'].fillna(data1['MPG_City'].mean(), inplace = True)
data1['MPG_Highway'].fillna(data1['MPG_Highway'].mean(), inplace = True)
data1['Weight'].fillna(data1['Weight'].mean(), inplace = True)
data1['Wheelbase'].fillna(data1['Wheelbase'].mean(), inplace = True)
data1['Length'].fillna(data1['Length'].mean(), inplace = True)
```
